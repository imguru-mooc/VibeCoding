## Step 2. 사다리 데이터 구조 만들기

### 📋 Claude 에 보낼 프롬프트

```
Python으로 사다리 게임의 가로선 데이터를 생성하는 함수를 만들어줘.

함수명: create_ladder(n_players, n_rows, rung_prob=0.4)

설명:
- n_players : 참가자 수 (세로선 개수)
- n_rows    : 사다리 높이 (가로선이 놓일 수 있는 행 수)
- rung_prob : 가로선 생성 확률 (기본값 0.4)

반환값: 2D bool 리스트 rungs[행][열]
- rungs[i][j] = True  →  j번 열과 j+1번 열 사이에 가로선 존재
- rungs[i][j] = False →  가로선 없음

규칙:
- 인접한 두 열에 동시에 가로선이 생기면 안 됨 (겹침 방지)

테스트:
- n_players=4, n_rows=5 로 실행
- 각 행과 가로선 위치를 출력
- 한국어 주석 포함
```

## Step 4. 경로 추적 알고리즘

### 📋 Claude 에 보낼 프롬프트

```
사다리 게임에서 특정 참가자의 이동 경로를 추적하는 Python 함수를 만들어줘.

함수명: trace_path(start_col, rungs)

매개변수:
- start_col : 출발 열 번호 (0부터 시작)
- rungs     : 2D bool 리스트 (Step 2에서 만든 가로선 데이터)

알고리즘:
1. 현재 열(col)에서 시작
2. 각 행을 순서대로 내려가면서:
   - 오른쪽 가로선 있으면 → col += 1
   - 왼쪽 가로선 있으면  → col -= 1
   - 없으면 그대로
3. 각 이동 후 col 값을 path 리스트에 기록

반환값: (최종_열_번호, 경로_리스트)

테스트:
- 4명 사다리에서 참가자 0~3번 각각의 경로 출력
- 출력 형식: "철수: [0, 1, 1, 0, 0] → 꽝!"

한국어 주석 포함
```
