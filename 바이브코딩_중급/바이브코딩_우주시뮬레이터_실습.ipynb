## 🌟 Step 3: 별이 빛나는 밤하늘 만들기

### 💬 AI에게 줄 프롬프트 예시
> *"Canvas에 랜덤하게 200개의 별을 찍어줘. 별마다 크기와 밝기가 다르게. 배경은 검은 우주."*

````python
# ============================================
# Step 3: Canvas로 별이 빛나는 밤하늘
# ============================================

html_code = """
<canvas id="space3" width="800" height="400" style="
    border-radius: 15px;
    display: block;
    margin: 0 auto;
"></canvas>

<script>
(function() {
    const canvas = document.getElementById('space3');
    const ctx = canvas.getContext('2d');

    // 배경: 우주
    const gradient = ctx.createRadialGradient(400, 200, 50, 400, 200, 400);
    gradient.addColorStop(0, '#1a0a3e');
    gradient.addColorStop(1, '#050510');
    ctx.fillStyle = gradient;
    ctx.fillRect(0, 0, 800, 400);

    // 별 200개 랜덤 생성
    for (let i = 0; i < 200; i++) {
        const x = Math.random() * 800;
        const y = Math.random() * 400;
        const size = Math.random() * 2.5 + 0.5;  // 0.5 ~ 3
        const brightness = Math.random() * 0.7 + 0.3;  // 0.3 ~ 1.0

        ctx.beginPath();
        ctx.arc(x, y, size, 0, Math.PI * 2);
        ctx.fillStyle = `rgba(255, 255, 255, ${brightness})`;
        ctx.fill();
    }

    // 라벨
    ctx.font = '14px Arial';
    ctx.fillStyle = '#556';
    ctx.fillText('🌟 Step 3: 별 200개가 랜덤으로 생성되었습니다', 20, 385);
})();
</script>
"""

display(HTML(html_code))
````

### ✏️ 직접 바꿔보기!
- 별 개수를 바꿔보세요: `200` → `500` (더 많이), `50` (더 적게)
- 별 최대 크기를 바꿔보세요: `2.5` → `5` (더 큰 별)
- 별 색상을 바꿔보세요: `255, 255, 255` → `200, 220, 255` (파란빛)
- **셀을 다시 실행할 때마다 별 위치가 달라집니다!** (랜덤)

## 🎬 Step 4: 별들을 움직여보자! (애니메이션)

### 💬 AI에게 줄 프롬프트 예시
> *"별들이 천천히 반짝이게 해줘. 그리고 화면 중앙에서 바깥으로 퍼져나가는 워프 효과도 넣어줘."*

````python
# ============================================
# Step 4: 애니메이션 - 스타 워프 효과!
# ============================================

html_code = """
<canvas id="space4" width="800" height="450" style="
    border-radius: 15px;
    display: block;
    margin: 0 auto;
    background: #000;
"></canvas>
<div style="text-align:center; color:#556; font-size:13px; margin-top:8px;">
    ⚡ 별들이 중앙에서 바깥으로 퍼져나갑니다 (워프 스피드!)
</div>

<script>
(function() {
    const canvas = document.getElementById('space4');
    const ctx = canvas.getContext('2d');
    const W = 800, H = 450;
    const cx = W / 2, cy = H / 2;

    // 별 데이터 생성
    const stars = [];
    for (let i = 0; i < 300; i++) {
        stars.push({
            x: (Math.random() - 0.5) * W,
            y: (Math.random() - 0.5) * H,
            z: Math.random() * W,
            pz: 0
        });
    }

    let speed = 8;  // 💡 이 값을 바꾸면 속도가 변합니다!

    function animate() {
        // 잔상 효과 (완전히 지우지 않고 반투명 덮기)
        ctx.fillStyle = 'rgba(0, 0, 10, 0.2)';
        ctx.fillRect(0, 0, W, H);

        for (let star of stars) {
            star.pz = star.z;
            star.z -= speed;

            // 화면 밖으로 나가면 리셋
            if (star.z <= 0) {
                star.x = (Math.random() - 0.5) * W;
                star.y = (Math.random() - 0.5) * H;
                star.z = W;
                star.pz = W;
            }

            // 원근 투영
            const sx = (star.x / star.z) * W + cx;
            const sy = (star.y / star.z) * H + cy;
            const px = (star.x / star.pz) * W + cx;
            const py = (star.y / star.pz) * H + cy;

            const size = Math.max(0, (1 - star.z / W) * 3);
            const brightness = Math.max(0, (1 - star.z / W));

            // 별의 궤적 (선)
            ctx.beginPath();
            ctx.moveTo(px, py);
            ctx.lineTo(sx, sy);
            ctx.strokeStyle = `rgba(180, 200, 255, ${brightness})`;
            ctx.lineWidth = size;
            ctx.stroke();

            // 별 점
            ctx.beginPath();
            ctx.arc(sx, sy, size * 0.8, 0, Math.PI * 2);
            ctx.fillStyle = `rgba(220, 240, 255, ${brightness})`;
            ctx.fill();
        }

        requestAnimationFrame(animate);
    }

    animate();
})();
</script>
"""

display(HTML(html_code))
````

### ✏️ 직접 바꿔보기!
- `speed = 8` → `speed = 2` (느리게) 또는 `speed = 20` (초고속 워프!)
- 별 개수: `300` → `600` (더 조밀하게)
- 잔상 효과: `0.2` → `0.05` (긴 꼬리) 또는 `0.5` (짧은 꼬리)
- 별 색상: `180, 200, 255` → `255, 180, 100` (따뜻한 톤)

## 🖱️ Step 5: 마우스로 조종하기!

### 💬 AI에게 줄 프롬프트 예시
> *"마우스를 움직이면 워프 방향이 바뀌게 해줘. 마우스 위치가 중심점이 되도록."*

````python
# ============================================
# Step 5: 마우스 인터랙션 추가!
# ============================================

html_code = """
<canvas id="space5" width="800" height="450" style="
    border-radius: 15px;
    display: block;
    margin: 0 auto;
    background: #000;
    cursor: crosshair;
"></canvas>
<div style="text-align:center; color:#88aacc; font-size:14px; margin-top:10px;">
    🖱️ <strong>마우스를 캔버스 위에서 움직여보세요!</strong> 워프 방향이 바뀝니다
</div>

<script>
(function() {
    const canvas = document.getElementById('space5');
    const ctx = canvas.getContext('2d');
    const W = 800, H = 450;

    // 마우스 위치 (기본: 중앙)
    let mouseX = W / 2, mouseY = H / 2;

    canvas.addEventListener('mousemove', (e) => {
        const rect = canvas.getBoundingClientRect();
        mouseX = (e.clientX - rect.left) * (W / rect.width);
        mouseY = (e.clientY - rect.top) * (H / rect.height);
    });

    // 별 데이터
    const stars = [];
    for (let i = 0; i < 400; i++) {
        stars.push({
            x: (Math.random() - 0.5) * W * 2,
            y: (Math.random() - 0.5) * H * 2,
            z: Math.random() * W,
            pz: 0
        });
    }

    let speed = 10;

    function animate() {
        ctx.fillStyle = 'rgba(0, 0, 8, 0.15)';
        ctx.fillRect(0, 0, W, H);

        for (let star of stars) {
            star.pz = star.z;
            star.z -= speed;

            if (star.z <= 0) {
                star.x = (Math.random() - 0.5) * W * 2;
                star.y = (Math.random() - 0.5) * H * 2;
                star.z = W;
                star.pz = W;
            }

            // 🔑 핵심 변경: 마우스 위치가 중심!
            const sx = (star.x / star.z) * W + mouseX;
            const sy = (star.y / star.z) * H + mouseY;
            const px = (star.x / star.pz) * W + mouseX;
            const py = (star.y / star.pz) * H + mouseY;

            const size = Math.max(0, (1 - star.z / W) * 3);
            const brightness = Math.max(0, (1 - star.z / W));

            // 색상: 거리에 따라 파랑→흰색
            const r = Math.floor(150 + 105 * brightness);
            const g = Math.floor(180 + 75 * brightness);
            const b = 255;

            ctx.beginPath();
            ctx.moveTo(px, py);
            ctx.lineTo(sx, sy);
            ctx.strokeStyle = `rgba(${r}, ${g}, ${b}, ${brightness})`;
            ctx.lineWidth = size;
            ctx.stroke();

            ctx.beginPath();
            ctx.arc(sx, sy, size * 0.8, 0, Math.PI * 2);
            ctx.fillStyle = `rgba(${r}, ${g}, ${b}, ${brightness})`;
            ctx.fill();
        }

        // 십자선 표시
        ctx.strokeStyle = 'rgba(0, 255, 255, 0.3)';
        ctx.lineWidth = 1;
        ctx.beginPath();
        ctx.arc(mouseX, mouseY, 15, 0, Math.PI * 2);
        ctx.stroke();

        requestAnimationFrame(animate);
    }

    animate();
})();
</script>
"""

display(HTML(html_code))
````

### ✏️ 직접 바꿔보기!
- 십자선 크기: `15` → `30` (더 큰 조준경)
- 별 색상을 무지개로: `b = 255` → `b = Math.floor(Math.random() * 255)`
- 클릭하면 속도가 바뀌게 하려면? → **Step 7에서 해봅니다!**

## 🪐 Step 6: 행성 궤도 추가하기

### 💬 AI에게 줄 프롬프트 예시
> *"중앙에 태양을 놓고, 3개의 행성이 각각 다른 속도와 색상으로 궤도를 돌게 해줘. 궤도 선도 보이게."*

````python
# ============================================
# Step 6: 태양계 - 행성 궤도 시뮬레이션
# ============================================

html_code = """
<canvas id="space6" width="800" height="500" style="
    border-radius: 15px;
    display: block;
    margin: 0 auto;
    background: #050510;
"></canvas>
<div style="text-align:center; color:#88aacc; font-size:14px; margin-top:10px;">
    🪐 3개의 행성이 태양 주위를 공전합니다
</div>

<script>
(function() {
    const canvas = document.getElementById('space6');
    const ctx = canvas.getContext('2d');
    const W = 800, H = 500;
    const cx = W / 2, cy = H / 2;

    // 배경 별
    const bgStars = [];
    for (let i = 0; i < 150; i++) {
        bgStars.push({
            x: Math.random() * W,
            y: Math.random() * H,
            size: Math.random() * 1.5 + 0.3,
            twinkle: Math.random() * Math.PI * 2
        });
    }

    // 행성 데이터 🔑 여기를 수정하면 행성이 바뀝니다!
    const planets = [
        { name: '수성', radius: 80,  size: 6,  speed: 0.03,  color: '#aaa',    angle: 0 },
        { name: '지구', radius: 150, size: 10, speed: 0.015, color: '#4488ff', angle: Math.PI / 3 },
        { name: '화성', radius: 210, size: 8,  speed: 0.008, color: '#ff6644', angle: Math.PI }
    ];

    let frame = 0;

    function animate() {
        ctx.clearRect(0, 0, W, H);

        // 배경 그라디언트
        const bg = ctx.createRadialGradient(cx, cy, 30, cx, cy, 400);
        bg.addColorStop(0, '#0a0a20');
        bg.addColorStop(1, '#020208');
        ctx.fillStyle = bg;
        ctx.fillRect(0, 0, W, H);

        // 배경 별 (반짝임)
        for (let s of bgStars) {
            const b = 0.4 + 0.6 * Math.sin(s.twinkle + frame * 0.02);
            ctx.beginPath();
            ctx.arc(s.x, s.y, s.size, 0, Math.PI * 2);
            ctx.fillStyle = `rgba(255, 255, 255, ${b})`;
            ctx.fill();
        }

        // 궤도선
        for (let p of planets) {
            ctx.beginPath();
            ctx.arc(cx, cy, p.radius, 0, Math.PI * 2);
            ctx.strokeStyle = 'rgba(255, 255, 255, 0.08)';
            ctx.lineWidth = 1;
            ctx.stroke();
        }

        // 태양 (글로우 효과)
        const sunGlow = ctx.createRadialGradient(cx, cy, 5, cx, cy, 50);
        sunGlow.addColorStop(0, '#ffee00');
        sunGlow.addColorStop(0.3, '#ffaa00');
        sunGlow.addColorStop(1, 'rgba(255, 100, 0, 0)');
        ctx.beginPath();
        ctx.arc(cx, cy, 50, 0, Math.PI * 2);
        ctx.fillStyle = sunGlow;
        ctx.fill();

        ctx.beginPath();
        ctx.arc(cx, cy, 18, 0, Math.PI * 2);
        ctx.fillStyle = '#ffdd00';
        ctx.fill();

        // 행성 이동 + 그리기
        for (let p of planets) {
            p.angle += p.speed;
            const px = cx + Math.cos(p.angle) * p.radius;
            const py = cy + Math.sin(p.angle) * p.radius;

            // 행성 글로우
            const glow = ctx.createRadialGradient(px, py, p.size * 0.5, px, py, p.size * 3);
            glow.addColorStop(0, p.color);
            glow.addColorStop(1, 'rgba(0,0,0,0)');
            ctx.beginPath();
            ctx.arc(px, py, p.size * 3, 0, Math.PI * 2);
            ctx.fillStyle = glow;
            ctx.fill();

            // 행성
            ctx.beginPath();
            ctx.arc(px, py, p.size, 0, Math.PI * 2);
            ctx.fillStyle = p.color;
            ctx.fill();

            // 이름 표시
            ctx.font = '12px Arial';
            ctx.fillStyle = 'rgba(255,255,255,0.6)';
            ctx.fillText(p.name, px + p.size + 5, py + 4);
        }

        frame++;
        requestAnimationFrame(animate);
    }

    animate();
})();
</script>
"""

display(HTML(html_code))
````

### ✏️ 직접 바꿔보기!
- 행성을 추가해보세요! `planets` 배열에 새 항목 추가:

## 🎮 Step 7: 최종 완성! 컨트롤 패널 추가

### 💬 AI에게 줄 프롬프트 예시
> *"워프 스피드를 조절하는 슬라이더, 별 색상을 바꾸는 버튼, 행성 궤도 보이기/숨기기 토글을 추가해줘. 컨트롤 패널은 캔버스 위에 투명하게 떠있도록."*

````python
# ============================================
# Step 7: 🎮 최종 완성판! 모든 기능 통합
# ============================================

html_code = """
<div id="container7" style="position:relative; width:800px; margin:0 auto;">

  <!-- 캔버스 -->
  <canvas id="space7" width="800" height="550" style="
      border-radius: 15px;
      display: block;
      background: #000;
      cursor: crosshair;
  "></canvas>

  <!-- 🎮 컨트롤 패널 -->
  <div id="controls7" style="
      position: absolute;
      top: 15px;
      left: 15px;
      background: rgba(0, 10, 30, 0.85);
      border: 1px solid rgba(0, 200, 255, 0.3);
      border-radius: 12px;
      padding: 15px 20px;
      color: #cde;
      font-family: 'Courier New', monospace;
      font-size: 13px;
      backdrop-filter: blur(5px);
      z-index: 10;
      min-width: 220px;
  ">
    <div style="color:#0ff; font-weight:bold; margin-bottom:12px; font-size:15px;">🎮 CONTROL PANEL</div>

    <label>⚡ 워프 스피드: <span id="speedVal7">8</span></label><br>
    <input type="range" id="speedSlider7" min="1" max="30" value="8" style="width:100%; margin:5px 0 12px;"><br>

    <label>🌟 별 개수: <span id="starVal7">300</span></label><br>
    <input type="range" id="starSlider7" min="50" max="800" value="300" step="50" style="width:100%; margin:5px 0 12px;"><br>

    <label>🎨 색상 테마:</label><br>
    <div style="display:flex; gap:6px; margin:8px 0 12px;">
      <button onclick="setTheme7('ice')" style="flex:1; padding:6px; border:1px solid #0af; background:rgba(0,100,255,0.3); color:#0af; border-radius:6px; cursor:pointer;">🧊 얼음</button>
      <button onclick="setTheme7('fire')" style="flex:1; padding:6px; border:1px solid #f60; background:rgba(255,100,0,0.3); color:#f60; border-radius:6px; cursor:pointer;">🔥 불꽃</button>
      <button onclick="setTheme7('rainbow')" style="flex:1; padding:6px; border:1px solid #f0f; background:rgba(200,0,200,0.3); color:#f0f; border-radius:6px; cursor:pointer;">🌈 무지개</button>
    </div>

    <label>🪐 행성 표시:</label><br>
    <div style="display:flex; gap:6px; margin:8px 0;">
      <button id="planetToggle7" onclick="togglePlanets7()" style="flex:1; padding:6px; border:1px solid #4f4; background:rgba(0,200,0,0.2); color:#4f4; border-radius:6px; cursor:pointer;">ON ✅</button>
      <button onclick="resetView7()" style="flex:1; padding:6px; border:1px solid #aaa; background:rgba(150,150,150,0.2); color:#aaa; border-radius:6px; cursor:pointer;">🔄 리셋</button>
    </div>

    <div style="margin-top:10px; color:#556; font-size:11px;">FPS: <span id="fps7">--</span></div>
  </div>

  <!-- 정보 표시 -->
  <div style="
      position: absolute;
      bottom: 15px;
      right: 15px;
      color: rgba(100,150,200,0.6);
      font-family: 'Courier New', monospace;
      font-size: 11px;
      text-align: right;
  ">
    🖱️ 마우스로 워프 방향 조종<br>
    🎮 슬라이더로 속도/개수 조절<br>
    🎨 테마 버튼으로 색상 변경
  </div>
</div>

<script>
(function() {
    const canvas = document.getElementById('space7');
    const ctx = canvas.getContext('2d');
    const W = 800, H = 550;

    // 상태 변수
    let mouseX = W / 2, mouseY = H / 2;
    let speed = 8;
    let starCount = 300;
    let showPlanets = true;
    let colorTheme = 'ice';  // 'ice', 'fire', 'rainbow'

    // 마우스 이벤트
    canvas.addEventListener('mousemove', (e) => {
        const rect = canvas.getBoundingClientRect();
        mouseX = (e.clientX - rect.left) * (W / rect.width);
        mouseY = (e.clientY - rect.top) * (H / rect.height);
    });

    // 별 생성
    let stars = [];
    function createStars(count) {
        stars = [];
        for (let i = 0; i < count; i++) {
            stars.push({
                x: (Math.random() - 0.5) * W * 2,
                y: (Math.random() - 0.5) * H * 2,
                z: Math.random() * W,
                pz: 0
            });
        }
    }
    createStars(starCount);

    // 행성
    const planets = [
        { name: '수성', radius: 60,  size: 5,  speed: 0.025, color: '#bbbbbb', angle: 0 },
        { name: '금성', radius: 100, size: 7,  speed: 0.018, color: '#ffcc44', angle: 1 },
        { name: '지구', radius: 150, size: 8,  speed: 0.012, color: '#4488ff', angle: 2 },
        { name: '화성', radius: 195, size: 6,  speed: 0.008, color: '#ff5533', angle: 3.5 },
    ];

    // 색상 테마
    function getStarColor(brightness, frame, index) {
        switch(colorTheme) {
            case 'fire':
                return `rgba(255, ${Math.floor(100 + 155 * brightness)}, ${Math.floor(50 * brightness)}, ${brightness})`;
            case 'rainbow':
                const hue = (index * 7 + frame * 0.5) % 360;
                return `hsla(${hue}, 100%, ${50 + 30 * brightness}%, ${brightness})`;
            default: // ice
                return `rgba(${Math.floor(150 + 105 * brightness)}, ${Math.floor(200 + 55 * brightness)}, 255, ${brightness})`;
        }
    }

    // 컨트롤 이벤트
    document.getElementById('speedSlider7').oninput = function() {
        speed = parseInt(this.value);
        document.getElementById('speedVal7').textContent = speed;
    };
    document.getElementById('starSlider7').oninput = function() {
        starCount = parseInt(this.value);
        document.getElementById('starVal7').textContent = starCount;
        createStars(starCount);
    };

    // 전역 함수
    window.setTheme7 = function(theme) { colorTheme = theme; };
    window.togglePlanets7 = function() {
        showPlanets = !showPlanets;
        const btn = document.getElementById('planetToggle7');
        btn.textContent = showPlanets ? 'ON ✅' : 'OFF ❌';
        btn.style.borderColor = showPlanets ? '#4f4' : '#f44';
        btn.style.color = showPlanets ? '#4f4' : '#f44';
    };
    window.resetView7 = function() {
        mouseX = W / 2; mouseY = H / 2;
        speed = 8;
        document.getElementById('speedSlider7').value = 8;
        document.getElementById('speedVal7').textContent = 8;
        colorTheme = 'ice';
    };

    // FPS 카운터
    let lastTime = performance.now(), frameCount = 0;
    let frame = 0;

    function animate() {
        // FPS 계산
        frameCount++;
        const now = performance.now();
        if (now - lastTime >= 1000) {
            document.getElementById('fps7').textContent = frameCount;
            frameCount = 0;
            lastTime = now;
        }

        // 배경
        ctx.fillStyle = 'rgba(0, 0, 5, 0.15)';
        ctx.fillRect(0, 0, W, H);

        // 워프 별
        for (let i = 0; i < stars.length; i++) {
            const star = stars[i];
            star.pz = star.z;
            star.z -= speed;

            if (star.z <= 0) {
                star.x = (Math.random() - 0.5) * W * 2;
                star.y = (Math.random() - 0.5) * H * 2;
                star.z = W;
                star.pz = W;
            }

            const sx = (star.x / star.z) * W + mouseX;
            const sy = (star.y / star.z) * H + mouseY;
            const px = (star.x / star.pz) * W + mouseX;
            const py = (star.y / star.pz) * H + mouseY;

            const size = Math.max(0, (1 - star.z / W) * 3.5);
            const brightness = Math.max(0, (1 - star.z / W));
            const color = getStarColor(brightness, frame, i);

            ctx.beginPath();
            ctx.moveTo(px, py);
            ctx.lineTo(sx, sy);
            ctx.strokeStyle = color;
            ctx.lineWidth = size;
            ctx.stroke();

            ctx.beginPath();
            ctx.arc(sx, sy, size * 0.7, 0, Math.PI * 2);
            ctx.fillStyle = color;
            ctx.fill();
        }

        // 행성 시스템
        if (showPlanets) {
            const pcx = W / 2, pcy = H / 2;

            // 궤도선
            for (let p of planets) {
                ctx.beginPath();
                ctx.arc(pcx, pcy, p.radius, 0, Math.PI * 2);
                ctx.strokeStyle = 'rgba(255, 255, 255, 0.06)';
                ctx.lineWidth = 1;
                ctx.stroke();
            }

            // 태양
            const sunGlow = ctx.createRadialGradient(pcx, pcy, 3, pcx, pcy, 35);
            sunGlow.addColorStop(0, 'rgba(255, 240, 0, 0.9)');
            sunGlow.addColorStop(0.4, 'rgba(255, 180, 0, 0.4)');
            sunGlow.addColorStop(1, 'rgba(255, 100, 0, 0)');
            ctx.beginPath();
            ctx.arc(pcx, pcy, 35, 0, Math.PI * 2);
            ctx.fillStyle = sunGlow;
            ctx.fill();
            ctx.beginPath();
            ctx.arc(pcx, pcy, 12, 0, Math.PI * 2);
            ctx.fillStyle = '#ffdd00';
            ctx.fill();

            // 행성
            for (let p of planets) {
                p.angle += p.speed;
                const ppx = pcx + Math.cos(p.angle) * p.radius;
                const ppy = pcy + Math.sin(p.angle) * p.radius;

                // 글로우
                const glow = ctx.createRadialGradient(ppx, ppy, p.size * 0.3, ppx, ppy, p.size * 2.5);
                glow.addColorStop(0, p.color);
                glow.addColorStop(1, 'rgba(0,0,0,0)');
                ctx.beginPath();
                ctx.arc(ppx, ppy, p.size * 2.5, 0, Math.PI * 2);
                ctx.fillStyle = glow;
                ctx.fill();

                ctx.beginPath();
                ctx.arc(ppx, ppy, p.size, 0, Math.PI * 2);
                ctx.fillStyle = p.color;
                ctx.fill();

                ctx.font = '11px Courier New';
                ctx.fillStyle = 'rgba(200,220,255,0.5)';
                ctx.fillText(p.name, ppx + p.size + 4, ppy + 3);
            }
        }

        // 십자선
        ctx.strokeStyle = 'rgba(0, 255, 255, 0.2)';
        ctx.lineWidth = 1;
        ctx.beginPath();
        ctx.arc(mouseX, mouseY, 12, 0, Math.PI * 2);
        ctx.stroke();
        ctx.beginPath();
        ctx.moveTo(mouseX - 18, mouseY);
        ctx.lineTo(mouseX + 18, mouseY);
        ctx.moveTo(mouseX, mouseY - 18);
        ctx.lineTo(mouseX, mouseY + 18);
        ctx.strokeStyle = 'rgba(0, 255, 255, 0.1)';
        ctx.stroke();

        frame++;
        requestAnimationFrame(animate);
    }

    animate();
})();
</script>
"""

display(HTML(html_code))
````
