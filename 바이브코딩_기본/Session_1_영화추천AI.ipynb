## 📌 Step 3: 콘텐츠 기반 추천 시스템 (20분)

### 🧠 핵심 개념 (30초 설명)

| 개념 | 설명 |
|------|------|
| **TF-IDF** | 단어의 중요도를 수치로 표현하는 방법. 자주 나오되 다른 문서엔 적은 단어가 높은 점수 |
| **코사인 유사도** | 두 벡터 사이의 각도로 유사성 측정. 1에 가까울수록 비슷함 |

### 💬 Claude에게 보낼 프롬프트:

```
위 영화 DataFrame(변수명: df)을 활용해서 콘텐츠 기반 영화 추천 시스템을 만들어줘.

구현 요구사항:
1. df의 '장르', '키워드', '감독' 컬럼을 결합하여 'features' 컬럼 생성
2. sklearn의 TfidfVectorizer로 TF-IDF 벡터화
3. cosine_similarity로 영화 간 유사도 매트릭스 생성
4. recommend_movies(title, n=5) 함수 구현:
   - 입력: 영화 제목, 추천 개수
   - 출력: 유사한 영화 n개 (제목, 장르, 평점, 감독, 유사도 점수 포함)
   - 존재하지 않는 영화 제목 입력 시 에러 처리
5. 추천 결과를 예쁘게 출력 (이모지, 박스 형태)

테스트:
- '기생충', '인터스텔라', '올드보이' 각각에 대해 추천 실행
- 유사도 히트맵(상위 15개 영화)도 그려줘
```
