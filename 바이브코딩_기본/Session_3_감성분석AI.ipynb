## 📌 Step 2: 리뷰 데이터 전처리 & TF-IDF (15분)

텍스트 데이터를 머신러닝이 이해할 수 있는 숫자로 변환합니다.

### 🧠 핵심 개념

| 단계 | 설명 |
|------|------|
| **토큰화** | 문장을 단어/형태소 단위로 분리 |
| **불용어 제거** | 의미 없는 단어(은, 는, 이, 가 등) 제거 |
| **TF-IDF** | 단어를 중요도 기반 숫자 벡터로 변환 |

### 💬 Claude에게 보낼 프롬프트:

```
감성 분류 모델을 위한 데이터를 준비해줘.

1. 리뷰 데이터 확장:
   - 긍정 리뷰 50개, 부정 리뷰 50개 (총 100개) 생성
   - 다양한 표현과 어투 사용 (존댓말, 반말, 이모티콘 등 혼합)
   - labels: 긍정=1, 부정=0

2. 한국어 토큰화 함수 구현:
   - Okt morphs(stem=True) 사용
   - 1글자 이하 단어 제거
   - 불용어 리스트 적용 (은, 는, 이, 가, 을, 를, 에, 의, 와, 과, 도, 로)

3. TF-IDF 벡터화:
   - sklearn TfidfVectorizer 사용
   - max_features=500

4. 학습/테스트 데이터 분리 (80:20)

5. 데이터 현황 출력:
   - 전체 리뷰 수, 긍정/부정 비율
   - TF-IDF 행렬 크기
   - 학습/테스트 데이터 크기
   - TF-IDF 상위 20개 특성(단어) 출력
```

### ⬇️ Claude가 생성한 코드를 아래 셀에 붙여넣고 실행하세요

## 📌 Step 3: 감성 분류 모델 3종 학습 & 비교 (20분)

3가지 머신러닝 모델을 학습시키고 성능을 비교합니다.

### 💬 Claude에게 보낼 프롬프트:

```
위에서 준비한 데이터(X_train, X_test, y_train, y_test)로
3가지 모델을 학습하고 비교해줘.

모델:
1. 로지스틱 회귀 (LogisticRegression)
2. 랜덤 포레스트 (RandomForestClassifier, n_estimators=100)
3. SVM (SVC, kernel='linear', probability=True)

각 모델에 대해:
- 학습 후 테스트 데이터로 예측
- 정확도(accuracy), 정밀도(precision), 재현율(recall), F1 스코어 계산
- classification_report 출력

결과 시각화 (2x2 서브플롯):
(1) 모델별 성능 지표 그룹 막대그래프 (정확도, F1, 정밀도, 재현율)
(2) 가장 좋은 모델의 Confusion Matrix 히트맵
(3) 모델별 학습 시간 비교 막대그래프
(4) 가장 좋은 모델에서 중요 특성(단어) Top 15 수평 막대그래프

가장 좋은 모델을 best_model 변수에 저장해줘.
한글 폰트 설정도 포함해줘.
```
