## 📌 Step 3: 카툰 & 유화 & 스케치 효과 (15분)

고급 이미지 필터로 사진을 예술 작품처럼 변환합니다!

### 🧠 핵심 개념

| 효과 | 원리 |
|------|------|
| **카툰화** | 엣지 검출 + 색상 양자화(bilateral filter)를 결합 |
| **유화 효과** | 픽셀 블록 단위로 가장 많은 색상을 선택(mode filter) |
| **연필 스케치** | 그레이스케일 반전 + 가우시안 블러 + 색상 나누기 |

### 💬 Claude에게 보낼 프롬프트:

1. cartoonize(img, num_bilateral=7, block_size=9)
   - 그레이스케일 → medianBlur → adaptiveThreshold로 엣지 마스크
   - bilateralFilter 반복 적용으로 색상 평탄화
   - 엣지 마스크와 결합

2. oil_painting(img, size=7, dynRatio=1)
   - cv2.xphoto.oilPainting 사용 (없으면 직접 구현)

3. pencil_sketch(img, blur_sigma=25)
   - 그레이스케일 → 반전 → 가우시안 블러 → divide blend
   - 흑백 스케치 + 컬러 스케치 둘 다 생성

2x2 서브플롯으로 원본 + 3가지 효과를 비교 표시해줘.
